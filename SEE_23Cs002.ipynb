```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/auto-mpg (1).csv')
df.head()
```

```python
x=df[['mpg']]
y=df[['weight']]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = (x, y, test_size=0.2, random_state=42)
x_train
x_test
y_train
y_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_square_error, mean_absolute_error, r2_score
mse=mean_square_error(y_test, y_pred)
mae=mean_absolute_error(y_test, y_pred)
r2=r2_score(y_test, y_pred)
rmse=np.sqrt(mse)

print("Mean Square Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
print("Root Mean Square Error:", rmse)
```
